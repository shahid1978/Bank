```python
import pandas as pd
df=pd.read_csv('bank.csv')
df
```

```python
df.columns
```

```python
df=df.drop(['balance', 'housing',
       'loan', 'contact', 'day', 'month', 'duration', 'campaign', 'pdays',
       'previous', 'poutcome', 'deposit'],axis=1)
df
```

```python
df.describe()
df.isnull().sum()
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
x=df.drop('default',axis=1)
y=df['default']

lb_xjob=LabelEncoder()
# for a in range(1,int(x.columns-1)):
#   x[a]=lb_x.fit_transform(x[a])
x['job']=lb_xjob.fit_transform(x['job'])
lb_xmar=LabelEncoder()
x['marital']=lb_xmar.fit_transform(x['marital'])
lb_xedu=LabelEncoder()
x['education']=lb_xedu.fit_transform(x['education'])
lb_y=LabelEncoder()
y=lb_y.fit_transform(y)
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
# x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
clf=KNeighborsClassifier(n_neighbors=4,p=1)
clf.fit(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
pred=clf.predict(x_test)
cm=confusion_matrix(y_test,pred)
cr=classification_report(y_test,pred)
```

```python
print(cr)
print(cm)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cm, annot=True, cmap="Blues", fmt="d", xticklabels=True, yticklabels=True)

plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
```

```python
from sklearn.model_selection import GridSearchCV
parameters={
              'n_neighbors':[x for x in range(1,15)]
           }
grid_search=GridSearchCV(clf,parameters,n_jobs=-1)
grid_search.fit(x_train,y_train)
best_model=grid_search.best_estimator_
best_model
```
